```python
# prompt: Unzip the two files in python and save them, they're test.ft.txt.bz2 and train.ft.txt.bz2

import bz2

def unzip_and_save(filename, new_filename):
  """Unzips a bz2 file and saves it to a new file."""
  with bz2.open(filename, 'rt') as infile, open(new_filename, 'w') as outfile:
    for line in infile:
      outfile.write(line)


unzip_and_save('test.ft.txt.bz2', 'test.ft.txt')
unzip_and_save('train.ft.txt.bz2', 'train.ft.txt')
```
